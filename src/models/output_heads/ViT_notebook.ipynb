```python
import torch
import torch.nn as nn

class ClassSpecificInputWeighting(nn.Module):
    
    def __init__(self):
        super(ClassSpecificInputWeighting, self).__init__()
        # Learnable alpha for each class
        self.alpha =  nn.Parameter(torch.tensor([0.5]))
        #self.fc = nn.Linear(embed_dim, num_classes)  # Final classifier

    def forward(self, tokens_tuple):
        (cls_token, image_tokens) = tokens_tuple

        # Use per-class alpha (reshaped for broadcasting)
        
        combined_representation = self.alpha * cls_token + (1 - self.alpha) * image_tokens

        return combined_representation

# Example usage
batch_size = 16
num_tokens = 197  # CLS + 196 patches
embed_dim = 128
num_classes = 5

vit_tokens = torch.rand(batch_size, num_tokens, embed_dim)  # Simulated ViT output

cls_token = vit_tokens[:, 0]  # CLS token
image_tokens = vit_tokens[:, 1:]  # Image tokens

# Compute mean of image tokens
image_representation = image_tokens.mean(dim=1)

# Initialize model
class_balancer = ClassSpecificInputWeighting()

inputs_tuple = (cls_token, image_representation)
# Get logits
logits = class_balancer(inputs_tuple)

print("Output Shape:", logits.shape)  # Expected: (batch_size, num_classes)
```
