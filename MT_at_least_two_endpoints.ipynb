```python
import pandas as pd
import os
```

```python
original_dir = r"/home/macraedc/data/MT_dataset/stratified_sampling_test_542.csv"
df = pd.read_csv(original_dir, sep=';')
```

```python
endpoint_columns = ['Aspiration_M06', 'Dysphagia_M06', 'Sticky_M06', 'Taste_M06', 'Xerostomia_M06']


# Count the number of values equal to -1 in the endpoint_columns, in each row
# then print the number of rows missing none, missing one, missing two, etc.
missing_counts_test = df[endpoint_columns].apply(lambda row: (row == -1).sum(), axis=1)
missing_summary_test = missing_counts_test.value_counts().sort_index()
print("Test set missingness summary:")
print(missing_summary_test)


# also print the rows missing 4 or more values
#print("Rows in test set missing 4 or more endpoint values:")
print(df[missing_counts_test == 4][endpoint_columns])
```

```python
df_at_least_two_endpoints = df[missing_counts_test != 4]
df_at_least_two_endpoints.to_csv(r"/home/macraedc/data/MT_dataset/stratified_sampling_test_542_at_least_two_endpoints.csv", sep=';', index=False)
```
