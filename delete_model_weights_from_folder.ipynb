```python
import os

model_file_name = "DlModel_Weights.pth"
```

```python
def delete_file_in_subfolders(directory, filename):
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file == filename:
                file_path = os.path.join(root, file)
                try:
                    os.remove(file_path)
                    #print(f"Deleted: {file_path}")
                except Exception as e:
                    print(f"Error deleting {file_path}: {e}")

# Example usage:

folder_path = '/home/macraedc/UQ_results/Data Deep Ensemble'  # Replace with your folder path

delete_file_in_subfolders(folder_path, model_file_name)
```
