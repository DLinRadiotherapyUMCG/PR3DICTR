```python
import numpy as np
import pandas as pd
from sklearn.model_selection import StratifiedShuffleSplit, ShuffleSplit

def stratified_cumulative_sampling(X, y, sample_sizes, step_size =10, random_state=42):
    """
    Perform stratified cumulative sampling.
    
    Parameters:
        X (array-like): Feature matrix.
        y (array-like): Target labels for stratification.
        sample_sizes (list): List of cumulative sample sizes (e.g., [100, 200, 300, ...]).
        random_state (int): Random seed for reproducibility.
    
    Returns:
        dict: Dictionary containing subsets with keys as sample sizes.
    """
    sampled_indices = set()
    subsets = {}

    for size in sample_sizes:
        print(size)
        remaining_indices = list(set(range(len(X))) - sampled_indices)
        
        strat_split = ShuffleSplit(n_splits=1, train_size=size - len(sampled_indices), random_state=random_state)
        new_indices, _ = next(strat_split.split(X[remaining_indices])) # , y[remaining_indices]))
        new_indices = np.array(remaining_indices)[new_indices]  # Convert to original index space
        
        sampled_indices.update(new_indices)
        subsets[size] = X[list(sampled_indices)]
    
    return subsets

# Example usage
np.random.seed(42)
X = np.random.rand(20, 3)  # 1100 samples, 10 features
y = np.random.randint(0, 1, 20)  # 3 classes for stratification

sample_sizes = list(range(3, 11, 1))
subsets = stratified_cumulative_sampling(X, y, sample_sizes)

# Checking sizes
for size, (X_subset) in subsets.items():
    print(f"Sample size {size}: {X_subset.shape}")
    print(X_subset)
```

```python
def perform_cumulative_sampling2(config, df, sampled_indices, target_sample_size):
    """
    """
    
    remaining_indices = list(set(range(len(df))) - set(sampled_indices))  # determine which patients have not already been sampled
    n_required_samples = target_sample_size - len(sampled_indices)  # how many more patients  we need to sample
    

    if n_required_samples == len(remaining_indices):
        new_indices = remaining_indices
    elif n_required_samples < 0:
        # if the number of required samples is less than 0, raise an error
        # i.e. there is not enough training data to sample the target number of patients
        raise ValueError(f"Number of required samples is less than 0. There is not enough training data to sample {target_sample_size} patients.")
    else:
        # shuffle the indicies of the remaining patients and select the required number of samples
        shuffle_split = ShuffleSplit(n_splits=1, train_size=n_required_samples) # , random_state=config['general']['seed'])
        new_indices, _ = next(shuffle_split.split(remaining_indices))
        new_indices = np.array(remaining_indices)[new_indices]  # Convert back to the original index space

    # update the list of sampled indices with the new_indices
    sampled_indices += list(new_indices)

    # check that the number of unique indices is equal to the target sample size
    assert len(sampled_indices) == len(set(sampled_indices)) == target_sample_size
    
    return df.iloc[sampled_indices], sampled_indices

np.random.seed(42)
X = np.random.rand(4, 3)  # 10 samples, 3 features
df = pd.DataFrame(X)

pseudo_config = {
    'general' : 
     {
        "experiment_name" : "TEST_experiment",
        "trialNumber" : "?",
        "seed" : 42,
    },
    "data" : {
        "sampled_indices" : [],
        #"n_training_patients" : 3
      },
    "paths" : {
        "results" : "directory"
        }
    }


data_amounts = [2, 3, 4]

pseudo_config["data"]['n_training_patients_sets'] = data_amounts

seeds = [42, 101]

datasets_list = []

for seed in seeds:
    #seed = i
    pseudo_config['general']['seed'] = seed
    pseudo_config["data"]['sampled_indices'] = []

    sampled_indices = []

    for data_amount in pseudo_config["data"]['n_training_patients_sets']:
        print(seed, data_amount)
        pseudo_config["data"]['n_training_patients'] = data_amount
        #pseudo_config["data"]['sampled_indices'] = []
        #print("----")
        
        sample, sampled_indices = perform_cumulative_sampling2(pseudo_config, df, sampled_indices, data_amount)
        print('main', sampled_indices)
        datasets_list.append(sample)

        

        #print(sample.head(10))
        
        print("----\n")
    print("XXXXXXXXXXXXXXXX\n")


    for fold_idx, D_set in enumerate(datasets_list):
        n_patients = pseudo_config["data"]['n_training_patients_sets'][fold_idx]
        pseudo_config["data"]['n_training_patients'] = n_patients

        experiment_dir = os.path.join(pseudo_config["paths"]["results"], pseudo_config["general"]["experiment_name"])
        folder_dir = os.path.join(experiment_dir, f"n_patients_{pseudo_config["data"]['n_training_patients_sets'][fold_idx]}", f"seed_{pseudo_config["general"]["seed"]}")
        print(folder_dir)
        print(len(D_set))

        #print(fold_idx, D_set)
        print("----\n")
```
