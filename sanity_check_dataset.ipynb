```python
import pandas as pd
```

```python
df = pd.read_csv('//home/macraedc/data/PRIMA_2025_stratified_sampling.csv', sep=';')
df.columns
```

```python
df['CT_Artefact'].value_counts()
```

```python
import numpy as np

# Map 0 -> 0, 2 -> 1, 1 -> random label (e.g., 'A', 'B', 'C')
random_labels = [0, 1]
def map_artefact(val):
    if val == 0:
        return 0
    elif val == 2:
        return 1
    elif val == 1:
        return np.random.choice(random_labels)
    else:
        return np.nan

df['UQ_Metal_Artefacts'] = df['CT_Artefact'].apply(map_artefact)


def map_artefact_with_missing_labels(val):
    if val == 0:
        return 0
    elif val == 2:
        return 1
    elif val == 1:
        return -1
    else:
        return np.nan

df['UQ_Metal_Artefacts_w_missing'] = df['CT_Artefact'].apply(map_artefact_with_missing_labels)
```

```python
df.to_csv('//home/macraedc/data/UQ_metal_artefact_dataset.csv', sep=';', index=False)
```
