```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

def plot_CT_dose(CT, rtdose):
    fig, axs = plt.subplots(3, 2, figsize=(5, 8))

    axs[0, 0].imshow(rtdose[0, 0, :, :], cmap='gray')
    axs[0, 0].set_title('RT Dose')
    axs[0, 0].axis('off')
    #fig.colorbar(axs[0, 0].images[0], ax=axs[0])

    #fig.colorbar(axs[0].images[0], ax=axs[0])
    axs[0, 1].imshow(CT[0, 0, :, :], cmap='gray')
    axs[0, 1].set_title('CT')
    axs[0, 1].axis('off')
    #fig.colorbar(axs[0, 1].images[0], ax=axs[1])

    axs[1, 0].imshow(rtdose[0, 50, :, :], cmap='gray')
    axs[1, 0].set_title('RT Dose')
    axs[1, 0].axis('off')

    axs[1, 1].imshow(CT[0, 50, :, :], cmap='gray')
    axs[1, 1].set_title('CT')
    axs[1, 1].axis('off')

    axs[2, 0].imshow(rtdose[0, -1, :, :], cmap='gray')
    axs[2, 0].set_title('RT Dose')
    axs[2, 0].axis('off')

    axs[2, 1].imshow(CT[0, -1, :, :], cmap='gray')
    axs[2, 1].set_title('CT')
    axs[2, 1].axis('off')

    plt.tight_layout()
    plt.show()
```

```python
path = r"\\zkh\appdata\RTDicom\Projectline_HNC_modelling\PRI2MA\MDACC\preprocessed\dataset_full_mdacc\1014175846\segmentation_map.npy"

rtdose = np.load(path)

CT = r"\\zkh\appdata\RTDicom\Projectline_HNC_modelling\PRI2MA\MDACC\preprocessed\dataset_full_mdacc\1014175846\ct.npy"
CT = np.load(CT)
CT = np.clip(CT, -200, 400)

print(rtdose.shape)
print(rtdose.max())

plot_CT_dose(CT, rtdose)
```
