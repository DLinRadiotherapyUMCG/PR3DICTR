```python
import os
import pandas as pd
```

```python
df = pd.read_csv("/home/macraedc/rt_pred_results/TTA/OS/model_1/all_predictions.csv", sep=';')

df
```

```python
import matplotlib.pyplot as plt

# Select columns with '_pred_' in their name
pred_cols = [col for col in df.columns if '_pred_' in col]

# Select first few rows (e.g., first 10)
df = df.sort_values(by=pred_cols[0], ascending=False)
subset = df[pred_cols].head(3)

# Plot each row as a separate histogram with a different color
colors = plt.cm.tab10.colors  # Use tab10 colormap for up to 10 distinct colors

plt.figure(figsize=(10, 6))
for idx, row in subset.iterrows():
    plt.hist(row.values, bins=20, alpha=0.5, color=colors[idx % len(colors)], label=f'Row {idx}')

plt.title("Spread of Prediction Values (First 10 Rows)")
plt.xlabel("Prediction Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```
