```python
from sklearn.metrics import roc_auc_score


# compute the AUCs of subgroups
def remove_missing_labels(df, label_col):
    """
    remove rows with missing labels from the dataframe. missing labels are equal to -1
    
    """

    df_temp = df.copy()
    df_temp = df_temp[df_temp[label_col] != -1]

    return df_temp

def compute_auc(df, label_col, pred_col):
    """
    compute the AUC of a given label and prediction column in the dataframe
    """
    # remove rows with missing labels
    df_temp = remove_missing_labels(df, label_col)

    # compute the AUC
    auc = roc_auc_score(df_temp[label_col], df_temp[pred_col])

    return auc
```

```python
# compute the AUCs of the dataframes, using the groupby function on the column "Technique"
endpoints = ["Aspiration_M06", 'Dysphagia_M06', 'Sticky_M06', 'Taste_M06', 'Xerostomia_M06']

column_name_group = "Sex"

for key, group in df_merged.groupby(column_name_group):
    result_dict = {}
    # compute the AUC of the group
    for endpoint in endpoints:
        pred_col_name = f"{endpoint}_pred"
        true_col_name = f"{endpoint}_true"

        auc = compute_auc(group, true_col_name, pred_col_name)

        result_dict[endpoint] = round(auc, 2)

    print(column_name_group, key, result_dict)

print("- "* 20)

if column_name_group == "Photons": column_name_group = "Protons"
elif column_name_group == "Smoking_1": column_name_group = "Roken"

for key, group in df_new_merged.groupby(column_name_group):
    result_dict = {}
    # compute the AUC of the group
    for endpoint in endpoints:
        pred_col_name = f"{endpoint}_pred"
        true_col_name = f"{endpoint}_true"

        auc = compute_auc(group, true_col_name, pred_col_name)

        result_dict[endpoint] = round(auc, 2)

    print(column_name_group, key, result_dict)
```
