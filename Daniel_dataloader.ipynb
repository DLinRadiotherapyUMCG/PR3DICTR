```python
def get_patient(dataloader, requested_patient_ID):
    patient_ID_list = dataloader.dataset.patient_IDs_list  # get the list of patientIDs in the dataloader

    requested_patient_idx = patient_ID_list.index(requested_patient_ID)  # get the index of the requested patient

    # get the data and target of the requested patient
    data = dataloader.dataset.__getitem__(requested_patient_idx)
    
    return data
        

get_patient(val_loader, "0079378")
```
