```python
df_dataset2 = pd.read_csv(r"C:\Users\S.P.M. de Vette\OneDrive - UMCG\Documents\DL_NTCP_Multitox\datasets\MT_dataset\stratified_sampling_test_542_2.csv", delimiter=';')
df_dataset2["PatientID"] = df_dataset2["PatientID"].astype(str).str.zfill(7)
df_dataset2 = df_dataset2[["PatientID", 'CT_Artefact', 'CT_Artefact_binarised']]
df_ens_preds = df_ens_preds.merge(df_dataset2, on="PatientID", how="left")
```
