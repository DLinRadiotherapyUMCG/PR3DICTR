```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class SpatialDropout3D(nn.Module):
    def __init__(self, drop_prob):
        super(SpatialDropout3D, self).__init__()
        self.drop_prob = drop_prob

    def forward(self, x):
        if not self.training or self.drop_prob == 0:
            return x

        # x: (N, C, D, H, W)
        # Create dropout mask with shape (N, C, 1, 1, 1)
        noise_shape = (x.size(0), x.size(1), 1, 1, 1)
        mask = x.new_empty(noise_shape).bernoulli_(1 - self.drop_prob)
        mask = mask / (1 - self.drop_prob)
        print(mask.shape)
        return x * mask
```

```python
import torch

print("PyTorch CUDA version:", torch.__version__)
print("Is CUDA available:", torch.cuda.is_available())
```

```python
x = torch.randn(1, 2, 4, 4, 4)

import matplotlib.pyplot as plt

mean_per_channel = x.mean(dim=[2, 3, 4])
print("Mean per channel before dropout:", mean_per_channel)

dropout = SpatialDropout3D(0.5)

dropout.train()
x_dropped = dropout(x)

mean_per_channel_dropped = x_dropped.mean(dim=[2, 3, 4])
print("Mean per channel after dropout:", mean_per_channel_dropped)
```
