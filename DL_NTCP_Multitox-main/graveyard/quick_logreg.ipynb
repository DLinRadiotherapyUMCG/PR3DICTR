## TASTE

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score

# Load the CSV file
dfX_path = os.path.join(per_tox_folder, 'Taste_M06_542.csv')
data = pd.read_csv(dfX_path, delimiter=';', index_col=0)

# Define the features and target
features = ['Parotid_meandose', 'OralCavity_Ext_meandose', 'Age'] # Replace with your actual feature column names
target = 'Taste_M06' # Replace with your actual target column name

# Split the data into training and testing sets based on the "Split" column
train_data = data[data['Split'] != 'test']
test_data = data[data['Split'] == 'test']

# Separate features and target for training and testing
X_train = train_data[features]
y_train = train_data[target]
X_test = test_data[features]
y_test = test_data[target]

# Create and train the logistic regression model
model = LogisticRegression(C=1e10)
model.fit(X_train, y_train)

# Make predictions on the test set
train_predictions = model.predict_proba(X_train)[:, 1]
test_predictions = model.predict_proba(X_test)[:, 1]

# Evaluate the model using AUC
train_auc = roc_auc_score(y_train, train_predictions)
print(f'Train AUC: {train_auc}')
test_auc = roc_auc_score(y_test, test_predictions)
print(f'Test AUC: {test_auc}')

# Print the coefficient weights
print("Coefficient weights:")
print(model.coef_)
print(model.intercept_)
```

# Xerostomia

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score

# Load the CSV file
dfX_path = os.path.join(per_tox_folder, 'Xerostomia_M06_542.csv')

data = pd.read_csv(dfX_path, delimiter=';', index_col=0)

# Define the features and target
features = ['Parotid_meandose', 'Submandibular_meandose', 'Xerostomia_W01_Een_beetje', 'Xerostomia_W01_Nogal_Heel_erg'] # Replace with your actual feature column names
subfeatures = [["Parotid_meandose", 'Xerostomia_W01_Een_beetje', 'Xerostomia_W01_Nogal_Heel_erg'], ["Submandibular_meandose", 'Xerostomia_W01_Een_beetje', 'Xerostomia_W01_Nogal_Heel_erg']]
target = 'Xerostomia_M06' # Replace with your actual target column name
nr_submodels = len(subfeatures)
# Split the data into training and testing sets based on the "Split" column
train_data = data[data['Split'] != 'test']
test_data = data[data['Split'] == 'test']

coeffs_dict_list = []
for feature_subset in subfeatures:
    # Separate features and target for training and testing
    X_train = train_data[feature_subset]
    y_train = train_data[target]
    X_test = test_data[feature_subset]
    y_test = test_data[target]

    # Create and train the logistic regression model
    model = LogisticRegression(C=1e10)
    model.fit(X_train, y_train)

    # Make predictions on the test set
    train_predictions = model.predict_proba(X_train)[:, 1]
    test_predictions = model.predict_proba(X_test)[:, 1]

    # Evaluate the model using AUC
    train_auc = roc_auc_score(y_train, train_predictions)
    print(f'Train AUC: {train_auc}')
    test_auc = roc_auc_score(y_test, test_predictions)
    print(f'Test AUC: {test_auc}')

    # Print the coefficient weights
    print("Coefficient weights:")
    print(model.coef_)
    print(model.intercept_)
    print("\n")

    keys = ['intercept'] + feature_subset
    values = np.append(model.intercept_, model.coef_)
    coeffs_i_dict = {k: v for (k, v) in zip(keys, values)}
    coeffs_dict_list.append(coeffs_i_dict)


```

```python
# Construct coefficients of final model
lr_coefficients = []
for f in ['intercept'] + features:
    coeff_i = 0
    # For-loop over coefficients of submodels
    for d_i in coeffs_dict_list:
        if f in d_i.keys():
            coeff_i += d_i[f]

    lr_coefficients.append(coeff_i / nr_submodels)


lr_coefficients
```
