```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt


from data_preproc.checks import plotting_functions
```

```python
df_mdacc_present = pd.read_csv('MDACC_features_present.csv')
mdacc_IDs = list(df_mdacc_present['PatientID'].astype(str).values)
len(mdacc_IDs)
```

```python
preprocessed_MDACC_patients = '//zkh/appdata/RTDicom/Projectline_HNC_modelling/PRI2MA/MDACC/preprocessed/dataset_full_mdacc'
preprocessed_MDACC_patients = list(os.listdir(preprocessed_MDACC_patients))
preprocessed_MDACC_patients
```

```python
mdacc_patients_preprocessed_and_all_features = list(set(mdacc_IDs) & set(preprocessed_MDACC_patients))
len(mdacc_patients_preprocessed_and_all_features)
```

# Plots !

```python
preprocessed_MDACC_patients = '//zkh/appdata/RTDicom/Projectline_HNC_modelling/PRI2MA/MDACC/preprocessed/dataset_full_mdacc'

figures_all_dir = "MDACC_figures_all"
os.makedirs(figures_all_dir, exist_ok=True)

for patient_id in mdacc_patients_preprocessed_and_all_features:
    patient_path = os.path.join(preprocessed_MDACC_patients, patient_id)

    #['ct.npy', 'rtdose.npy', 'segmentation_map.npy']
    ct_arr = np.load(os.path.join(patient_path, 'ct.npy'))[0]
    rtdose_arr = np.load(os.path.join(patient_path, 'rtdose.npy'))[0]
    segmentation_map_arr = np.load(os.path.join(patient_path, 'segmentation_map.npy'))[0]

    plotting_rows_dicts = [
        {
            "Label": "CT,\nDose,\nSegmentation",
            "CT": ct_arr,
            "RTDOSE": rtdose_arr,
            "RTSTRUCT": segmentation_map_arr
        },
        {
            "Label": "CT",
            "CT": ct_arr,
        },
        {
            "Label": "Dose",
            "RTDOSE": rtdose_arr,
        },
        {
            "Label": "Segmentation",
            "RTSTRUCT": segmentation_map_arr
        }
    ]

    slices = [10, 20, 30, 40, 50, 60, 70, 80, 90]
    fig, axes = plotting_functions.plot_slices(plotting_rows_dicts, slices)# , title=f"{patient_id} slices")

    #figures_folder_img_dir = os.path.join(figures_dir, patient_id, 'all.png')
    #fig.savefig(figures_folder_img_dir)
    figures_all_folder_img_dir = os.path.join(figures_all_dir, '{}.png'.format(patient_id))
    fig.savefig(figures_all_folder_img_dir)
    plt.close(fig)
    
    #print(patient, patient_files)
```
