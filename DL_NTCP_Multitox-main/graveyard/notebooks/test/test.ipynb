```python
import torch
import numpy as np
import pandas as pd
```

```python
patient_ids = ['abc', 'def']
mode_list = ['train', 'val']
y_pred_dict = {'xer': [[0.1, 0.9], [0.1, 0.9]],
              'dys': [[0.05, 0.95], [0.3, 0.7]],
              'taste': [[0.6, 0.4], [0.2, 0.8]]}
y_true_dict = {'xer': [[1, 0], [0, 1]],
              'dys': [[0, 1], [0, 1]],
              'taste': [[1, 0], [1, 0]]}
endpoint_list = ['xer', 'dys', 'taste']


df_patient_ids = pd.DataFrame(patient_ids, columns=['PatientID'])
df_mode = pd.DataFrame(mode_list, columns=['Mode'])
df_y = pd.concat([df_patient_ids, df_mode], axis=1)

for endpoint in endpoint_list:
    y_pred = y_pred_dict[endpoint]
    y_true = y_true_dict[endpoint]
    
    y_pred = [np.array(x) for x in y_pred]
    y_true = [np.array(x) for x in y_true]
    
    num_cols = y_pred[0].shape[0]
    
    df_y_pred = pd.DataFrame(y_pred, columns=['{}_pred_{}'.format(endpoint, c) for c in range(num_cols)])
    df_y_true = pd.DataFrame(y_true, columns=['{}_true_{}'.format(endpoint, c) for c in range(num_cols)])
    df_y = pd.concat([df_y, df_y_pred, df_y_true], axis=1)
    
```
