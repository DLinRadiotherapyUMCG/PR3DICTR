```python
import os
import copy
import torch
import random
import numpy as np
import pandas as pd
endpoint_list = ['Xer', 'Dys', 'Taste']
nr_epochs_not_improved_dict = {endpoint: 0 for endpoint in endpoint_list}
patience = 7
```

```python
best_val_avg_auc_value_dict = {endpoint: -1 for endpoint in endpoint_list}
best_val_avg_auc_value_dict
```

```python
random.seed(0)
val_auc_value_dict = {endpoint: random.random() for endpoint in endpoint_list}
val_auc_value_dict  # dict of values, NOT dict of list of values
```

```python
nr_of_endpoints_improved = 0
for endpoint in endpoint_list:
    nr_of_endpoints_improved += (val_auc_value_dict[endpoint] > best_val_avg_auc_value_dict[endpoint])
```

```python
endpoint_improved_dict = dict()
for endpoint in endpoint_list:
    endpoint_improved_dict[endpoint] = (val_auc_value_dict[endpoint] > best_val_avg_auc_value_dict[endpoint])
    if endpoint_improved_dict[endpoint]:
        nr_epochs_not_improved_dict[endpoint] = 0
    else:
        nr_epochs_not_improved_dict[endpoint] += 1
```
