```python
from tabulate import tabulate
import config

class FakeLogger():
    def __init__(self):

        self.dict_names = []
        self.auc_dicts_list = []

        self.best_auc_dict = None
        self.best_epoch = None
        self.current_epoch = None


    def add_epoch_dict(self, name, auc_dict, loss, avg_auc, epoch_n):
        self.current_epoch = epoch_n
        self.dict_names.append(name)
        #auc_dict = auc_dict.copy()
        auc_dict['Average AUC'] = avg_auc
        auc_dict['Loss'] = loss
        self.auc_dicts_list.append(auc_dict)

    def set_best_epoch_dict(self, best_auc_dict, loss, avg_auc, epoch_n):
        self.best_epoch = epoch_n
        best_auc_dict['Average AUC'] = avg_auc
        best_auc_dict['Loss'] = loss
        self.best_auc_dict = best_auc_dict

    def print_epoch_table(self, mode='epoch'):
        #decimals = config.nr_of_decimals
        auc_dicts_list = self.auc_dicts_list
        if self.best_auc_dict is not None and mode == 'epoch':
            auc_dicts_list.append(self.best_auc_dict)
            self.dict_names.append('Best Validation')
            
        names = self.dict_names

        keys = list(set().union(*auc_dicts_list))
        keys.sort()


        # Move "Loss" and "Average AUC" to the start of the list
        keys.remove("Loss")
        keys.remove("Average AUC")
        keys = ["Average AUC", "Loss"] + keys

        header = [''] + keys
        header = [h.replace('_', ' ').title() for h in header] # replace _ with space and capitalize
        rows = []
        
        for name, d in zip(names, auc_dicts_list):
            #print(d)
            row_values = []
            for key in keys:
                row_values.append(f"{d[key]:.3f}" if key in d else "N/A")
            row = [name] + row_values
            rows.append(row)

        # pad the strings to be equal length
        # first find the max length of each column
        max_length = max([len(h) for h in header])

        # pad all of the items in the rows and header to be the same length as the longest item
        header = [h.center(max_length) for h in header]
        rows = [[value.rjust(max_length) for value in row] for row in rows]

        table = tabulate(rows, headers=header, tablefmt='pretty')
        #table = table.split('\n')
        print(f"Current epoch:    {self.current_epoch}")
        print(f"Best (val) epoch: {self.best_epoch}")
        print(table)

        # delete the epochs we've just printed
        self.dict_names = []
        self.auc_dicts_list = []
        






train_dict = {'Xerostomia': 1, 'Sticky': 2, 'Taste': 3, 'Aspiration': 3}
val_dict =   {'Xerostomia': 1, 'Sticky': 4, 'Taste': 6, 'Aspiration': 3}
best_val =  {'Xerostomia': 10.1343, 'Sticky': 8.1, 'Taste': 9.232, 'Aspiration': 3}

logger = FakeLogger()

logger.add_epoch_dict('Train', train_dict, loss=0.1, avg_auc=0.9, epoch_n=1)
logger.add_epoch_dict('Validation', val_dict, loss=0.4, avg_auc=0.7, epoch_n=1)

logger.set_best_epoch_dict(best_val, loss=0.5, avg_auc=0.7, epoch_n=5)

logger.print_epoch_table()
```

```python

def print_table(*dicts):
    # Get the keys from the dictionaries
    keys = list(set().union(*dicts))
    keys.sort()

    #keys = list(keys)
    print(keys)

    # Print header
    header = "\t".join("{}".format(key) for key in keys)
    #print(header)

    names = ["Train", "Val", "Test"]
    
    # Print values
    rows = []

    for name, d in zip(names, dicts):
        row = [name] + [d[key] for key in keys]
        #row = "\t".join(str(d.get(key, '')) for d in dicts)
        #print(row)
        rows.append(row)

    table = tabulate(rows, headers=keys, tablefmt="pretty")
    print(table)

# Example dictionaries
train_dict = {'Xero': 1, 'Sticky': 2, 'Taste': 3}
val_dict =   {'Xero': 1, 'Sticky': 4, 'Taste': 6}
test_dict =  {'Xero': 1, 'Sticky': 8, 'Taste': 9}

# Pivot the table
print_table(train_dict, val_dict, test_dict)
```
