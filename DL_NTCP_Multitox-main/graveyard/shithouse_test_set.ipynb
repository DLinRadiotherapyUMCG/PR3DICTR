```python
import pandas as pd
import numpy as np


df1_path = 'datasets\dataset_dan\stratified_sampling_test_101.csv'
df2_path = 'datasets\dataset_dan\stratified_sampling_test_101_NEW.csv'

df1 = pd.read_csv(df1_path, delimiter=';', index_col=None)
```

```python
import pandas as pd

# Step 1: Import the CSV files into two separate dataframes
df1 = pd.read_csv(df1_path, delimiter=';', index_col=None)
df2 = pd.read_csv(df2_path, delimiter=';', index_col=None)

# Step 2: Filter dataframe 1 to only include rows where `Split=='test'`
old_test_patients = df1[df1['Split'] == 'test']

# overwrite the labels
df2["Split"] = "train_val"

# Step 3: Update the `PatientsID` in dataframe 2 to 'test' where the `PatientsID` matches those in the filtered dataframe 1
df2.loc[df2['PatientID'].isin(old_test_patients['PatientID']), 'Split'] = 'test'

# Optionally, save the updated dataframe 2 to a new CSV file
df2.to_csv('updated_dataframe2.csv', sep=';', index=False)
```
