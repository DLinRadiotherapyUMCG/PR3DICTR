```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from lifelines import KaplanMeierFitter
```

```python
df_OS_data = pd.read_csv("/home/macraedc/rt_pred_results/OS test/Trial_4/KFold1/model_predictions.csv", sep=';', index_col='PatientID')
df_OS_data

df_train_val = df_OS_data[df_OS_data['Mode'] == 'val']
df_test = df_OS_data[df_OS_data['Mode'] == 'test']
```

```python
df_train_val
```

```python
T = df_train_val['OS_months_true'].values
E = df_train_val['OS_event_true'].values
preds = df_train_val['OS_pred'].values

kmf = KaplanMeierFitter()
ax = kmf.fit(T, E, label="kmf.plot_survival_function()")

kmf.plot_survival_function()
```

```python
preds_dict = {"OS" : df_train_val["OS_pred"].values}

labels_dict = {"OS": np.stack([df_train_val["OS_event_true"].values, df_train_val["OS_months_true"].values], axis=1)}
```

```python
def make_KM_plot(preds_dict, labels_dict):
    # seperate the predictions into a high risk and low risk group, using the median prediction as the threshold
    median_pred = np.median(preds_dict["OS"])
    lowrisk_group_list = np.where(preds_dict["OS"] < median_pred)[0]
    highrisk_group_list = np.where(preds_dict["OS"] >= median_pred)[0]

    event_labels = labels_dict["OS"][:, 0]
    time_labels = labels_dict["OS"][:, 1]

    print(len(lowrisk_group_list))

    print(preds_dict["OS"][lowrisk_group_list].shape, preds_dict["OS"][highrisk_group_list].shape)


    
    plt.figure()
    ax = plt.subplot(111)

    kmf_0 = KaplanMeierFitter()
    ax = kmf_0.fit(time_labels[lowrisk_group_list], event_labels[lowrisk_group_list], label='Low risk').plot_survival_function(ax=ax, show_censors=True, ci_show=True)
    kmf_1 = KaplanMeierFitter()
    ax = kmf_1.fit(time_labels[highrisk_group_list], event_labels[highrisk_group_list], label='High risk').plot_survival_function(ax=ax, show_censors=True, ci_show=True)
    from lifelines.plotting import add_at_risk_counts
    add_at_risk_counts(kmf_0, kmf_1, ax=ax, fontsize=14, rows_to_show=['At risk'])
    #add_at_risk_counts(kmf_1, ax=ax, fontsize=14, rows_to_show=['At risk']) 
    ax.set_xlim(0,60)
    ax.set_ylim(0.0, 1.0)
    ax.set_ylabel("OS rate", fontsize = 16, fontweight = 'bold') # new
    ax.set_xlabel('Time (months)',fontsize = 16, fontweight = 'bold')
    ax.legend( loc='lower left', prop={ 'size': 14})
        
    ax.tick_params(axis='both',labelsize = 14)  

    plt.tight_layout()
    plt.show()

make_KM_plot(preds_dict, labels_dict)
```
